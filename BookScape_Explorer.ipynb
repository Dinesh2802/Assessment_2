```python
# Assuming data['items'] contains the list of books fetched from the API
if 'items' in data:
    for i in range(min(40, len(data['items']))):  # Loop through the first 40 books or fewer if there aren't that many
        book = data['items'][i]  # Access the book at index i
        volume_info = book.get('volumeInfo', {})  # Get volume info of the book
        sale_info = book.get('saleInfo', {})  # Get sale info of the book

        # Access book details
        book_id = book.get('id', '')
        book_selfLink = book.get('selfLink', '')  # Corrected this line
        book_title = volume_info.get('title', '')
        book_sub_title = volume_info.get('subtitle', '')
        book_authors = ', '.join(volume_info.get('authors', []))  # Join authors if there are multiple
        book_description = volume_info.get('description', '')
        book_industryIdentifiers = volume_info.get('industryIdentifiers', [{}])[0].get('identifier', '')

        # Corrected these lines to properly access readingModes
        book_readingModesText = volume_info.get('readingModes', {}).get('text', False)
        book_readingModesImage = volume_info.get('readingModes', {}).get('image', False)

        book_page_count = volume_info.get('pageCount', 0)
        book_categories = ', '.join(volume_info.get('categories', []))  # Join categories if there are multiple
        book_language = volume_info.get('language', '')
        book_image_link = volume_info.get('imageLinks', {}).get('smallThumbnail', '')
        book_ratings_count = volume_info.get('ratingsCount', 0)
        book_average_rating = volume_info.get('averageRating', 0)

        # Access sale info
        book_saleInfo = sale_info.get('country', '')
        book_saleability = sale_info.get('saleability', '')
        book_is_ebook = sale_info.get('isEbook', False)

        # Correctly accessing price information
        book_listPriceAmount = sale_info.get('listPrice', {}).get('amount', 0)
        book_listPriceCurrency = sale_info.get('listPrice', {}).get('currencyCode', '')
        book_retailPriceAmount = sale_info.get('retailPrice', {}).get('amount', 0)
        book_retailPriceCurrency = sale_info.get('retailPrice', {}).get('currencyCode', '')

        book_buy_link = sale_info.get('buyLink', '')
        book_published_date = volume_info.get('publishedDate', '')
        book_publisher = volume_info.get('publisher', '')

        # Print book information
        print(f"Book {i + 1}:")
        print(f"Book ID: {book_id}")
        print(f"Self Link: {book_selfLink}")
        print(f"Title: {book_title}")
        print(f"Subtitle: {book_sub_title}")
        print(f"Authors: {book_authors}")
        print(f"Description: {book_description[:200]}...")  # Limiting description to first 200 chars
        print(f"Industry Identifier: {book_industryIdentifiers}")
        print(f"Reading Modes Text: {book_readingModesText}")
        print(f"Reading Modes Image: {book_readingModesImage}")
        print(f"Page Count: {book_page_count}")
        print(f"Categories: {book_categories}")
        print(f"Language: {book_language}")
        print(f"Image Link: {book_image_link}")
        print(f"Ratings Count: {book_ratings_count}")
        print(f"Average Rating: {book_average_rating}")
        print(f"Sale Info Country: {book_saleInfo}")
        print(f"Saleability: {book_saleability}")
        print(f"Is eBook: {'True' if book_is_ebook else 'False'}")
        print(f"List Price Amount: {book_listPriceAmount}")
        print(f"List Price Currency: {book_listPriceCurrency}")
        print(f"Retail Price Amount: {book_retailPriceAmount}")
        print(f"Retail Price Currency: {book_retailPriceCurrency}")
        print(f"Buy Link: {book_buy_link}")
        print(f"Published Date: {book_published_date}")
        print(f"Publisher: {book_publisher}")
        print("-" * 40)  # Separator for readability

else:
        print("No books found in the response.")

```
